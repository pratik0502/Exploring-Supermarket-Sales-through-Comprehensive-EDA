# EDA of SuperMarket Sales

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.heatmap(df.corr(),annot=True);
```

**1.Max correlation is between profit and sales**

```python
plt.figure(figsize=(12,5))
sns.countplot(x=df['Ship Mode'],order=df['Ship Mode'].value_counts().index.sort_values())
plt.xlabel('Ship Mode',fontsize=12)
plt.ylabel('Counts',fontsize=12)
plt.title('Counts Vs Ship Mode',fontsize=16);
```

                            More of the deliveries were done by Standard Class

```python
plt.figure(figsize=(12,5))
plt.pie(df['Segment'].value_counts(),labels=df['Segment'].value_counts().index,autopct='%1.2f%%');
```

More orders were placed by Consumer, followed by Corporate and Home Office

```python
plt.figure(figsize=(12,4))
sns.countplot(y=df['State'],order=df['State'].value_counts().index[0:5],palette='magma')
plt.xlabel('Orders',fontsize=14)
plt.ylabel('States',fontsize=14)
plt.title('States Vs Orders',fontsize=16);
```

            More orders were from California, followed by New York and Texas

```python
plt.figure(figsize=(10,4))
sns.countplot(x=df['Category'],order=df['Category'].value_counts().index.sort_values(),palette="mako")
plt.xlabel('Category',fontsize=12)
plt.ylabel('Orders',fontsize=12)
plt.title('Orders Vs Category',fontsize=16);
```

                    Most of the ordered items was of  Office Supplies 

```python
plt.figure(figsize=(12,5))
sns.countplot(y=df['Sub-Category'],order=df['Sub-Category'].value_counts().index[0:10],palette='crest')
plt.xlabel('Orders',fontsize=12)
plt.ylabel('Sub-Category',fontsize=12)
plt.title('Orders Vs Sub-Category',fontsize=16);
```

                            max order were of binders, followed by paper and furnishing

```python
plt.figure(figsize=(13,5))
sns.barplot(x=df["Category"],y=df["Sales"],palette='rocket',ci=None)
# plt.xticks(rotation='vertical')
plt.xlabel('Category',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.title('Sales Vs Category',fontsize=16);
```

                    More sales were of technology, followed by furniture and office supplies

```python
plt.figure(figsize=(13,5))
sns.barplot(y=df["Region"],x=df["Sales"],hue=df["Category"],palette='PuBuGn_r',ci=None)
plt.xlabel('Sales',fontsize=12)
plt.ylabel('Rregion',fontsize=12)
plt.title('Sales Vs Region',fontsize=16);
```

            More Technology products were sold in South, followed by East,West and Central Region

```python
plt.figure(figsize=(13,5))
sns.barplot(x=df["Profit"],y=df["Region"],hue=df['Category'],ci=None,palette="magma")
plt.xlabel('Profit',fontsize=12)
plt.ylabel('Region',fontsize=12)
plt.title('Region Vs Profit',fontsize=16);
```

Even though more items were sold in South Region, But more profit was earned from East region,followed Central,West and South


There was also a loss from Central region in Furniture Products

```python
plt.figure(figsize=(13,5))
sns.barplot(x=df['Segment'],y=df['Sales'],hue=df['Category'],ci=None,palette='Paired')
plt.xlabel('Segment',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.title('Segment Vs Sales',fontsize=16);
```

More technology products were Sales from Home office, followed by corporate and consumer.

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df['Segment'],y=df['Profit'],hue=df['Category'],ci=None,palette='Paired')
plt.xlabel('Segment',fontsize=12)
plt.ylabel('Profit',fontsize=12)
plt.title('Segment VS profit',fontsize=16);
```

As Sales was more from Home Office the profit was also more from Home Office segment, followed by Corporate and Consumer

```python
plt.figure(figsize=(10,14))
sns.barplot(y=df["State"],x=df["Sales"],hue=df["Category"],ci=None,palette="rocket")
plt.xlabel('Sales',fontsize=20)
plt.ylabel('States',fontsize=20)
plt.title('Sales Vs Sates',fontsize=24);
```

More Furniture products was sold in Vermont State,followed by Wyoming.

More technology products was sold in District in columbia, followed by Rhode island.

More Office supplies was sold in Vermont , followed by Nevada

```python
plt.figure(figsize=(10,14))
sns.barplot(y=df["State"],x=df["Profit"],hue=df["Category"],ci=None,palette="rocket")
plt.xlabel('Profit',fontsize=20)
plt.ylabel('States',fontsize=20)
plt.title('Profit Vs Sates',fontsize=24);
```

More Profit was made in District of columbia from Technology Category

More Profit was made in Vermont from Furniture Category

More Profit was made in Minnesota of columbia From office supplies category

More Loss was made in Ohio from Technology Category

More Loss was made in North Carolina from Furniture Category
